```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,4),dpi = 200)
sns.scatterplot(x='salary',y='sales',data = df,hue='level of education',
                palette='Dark2',style='level of education')
```

```python
plt.figure(figsize=(5,8),dpi=200)
sns.rugplot(x='salary',data = df,height = 0.5)
```

```python
#sns.set(style='darkgrid')
sns.displot(data=df,x='salary',bins=10,color='red',edgecolor='blue',
            linewidth=4)
```

```python
sns.histplot(data=df,x='salary',kde=True)
```

```python
sns.kdeplot(data=df,x='salary')
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='level of education',hue='division')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(data=df,x='level of education',y='salary',estimator=np.mean,ci='sd')
```

```python
df = pd.read_csv('/content/StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df,y='math score',x='test preparation course')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df,y='reading score',x='parental level of education',
            hue = 'test preparation course')

plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.violinplot(data=df,x='reading score',y='parental level of education',inner='quartile')
```

```python
sns.swarmplot(data=df,x='math score',y='gender',size = 2)
```

```python
sns.jointplot(data=df,x='math score',y = 'reading score',hue='gender')
```

```python
sns.pairplot(data=df,hue='gender')
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',
            col='lunch',row='test preparation course')
```
